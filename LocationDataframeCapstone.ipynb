```python
#importing beautifulsoup
from bs4 import BeautifulSoup
```

```python
#getting the rquired html page
import requests
req=requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
page=req.text
soup=BeautifulSoup(page,'html.parser')
```

```python
soup
```

```python
print(soup.prettify())
```

```python
#setting the table to a new variable
html_soup=soup.find('table')
type(html_soup)
```

```python
#getting rows of the table
rows=html_soup.find_all('tr')
type(rows)
```

```python
#getting column names as a list
columns=html_soup.find_all('th')[:]

for i in range(len(columns)):
    columns[i]=columns[i].get_text()
    #x=x.get_text()
    columns[i]=columns[i].replace('<th>','')
    columns[i]=columns[i].replace('</th>','')
    columns[i]=columns[i].replace('\n','')
#columns.get_text().replace('<th>','')
```
