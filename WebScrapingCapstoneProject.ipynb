# Web Scraping the Postal Code Data of Toronto City

```python
#importing BeautifulSoup library
from bs4 import BeautifulSoup
```

```python
#getting html of required website and converting it as a BeautifulSoup object
import requests
req=requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
page=req.text
soup=BeautifulSoup(page,'html.parser')
```

```python
soup
```

```python
#prettifying the html text
print(soup.prettify())
```

```python
#findinf the required table and setting it to a new variable
html_soup=soup.find('table')
type(html_soup)
```

```python
#getting the rows of the table
rows=html_soup.find_all('tr')
type(rows)
```

```python
#getting column names as a list
columns=html_soup.find_all('th')[:]

for i in range(len(columns)):
    columns[i]=columns[i].get_text()
    #x=x.get_text()
    columns[i]=columns[i].replace('<th>','')
    columns[i]=columns[i].replace('</th>','')
    columns[i]=columns[i].replace('\n','')
#columns.get_text().replace('<th>','')
```

```python
columns
```

```python
rows
```

```python
#getting row values as a list
row_values=html_soup.find_all('td')[:]
for i in range(len(row_values)):
    row_values[i]=row_values[i].get_text()
    row_values[i]=row_values[i].replace('<td>','')
    row_values[i]=row_values[i].replace('</td>','')
    row_values[i]=row_values[i].replace('\n','')
```

```python
#grouping the rows according to no. of columns
N = 3
subList = [row_values[i:i+3] for i in range(0, len(row_values), 3)]
subList
```

Now, aligning everything to get the desired Dataframe

```python
import pandas as pd
df=pd.DataFrame(data=subList,index=None,columns=columns)
```

```python
df
```

```python
#dropping rows with non-assigned borough values
df.drop(df[df.Borough=='Not assigned'].index,axis=0,inplace=True)
```

```python
df.reset_index(inplace=True)
df
```

```python
for i in range(len(df.Neighbourhood)):
    df.Neighbourhood[i]=df.Neighbourhood[i].replace('/',',')
#df.Neighborhood[2].replace('/',',')
df.drop('index',axis=1,inplace=True)
```
